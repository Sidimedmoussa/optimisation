```python
from ultralytics import YOLO
```

```python
import cv2
import pygame

model = YOLO("best.pt") 

cap = cv2.VideoCapture(0)

raisepygame.init()
pygame.mixer.init()
fire_sound = pygame.mixer.Sound("sound.mp3")

if not cap.isOpened():
    print("Error: Unable to open camera.")
    exit()
else:
    print("done......")

# Looping to capture frames from the camera
    
while True:
    # Captur fram by frame
    ret, frame = cap.read()
    


    # Display the frame
    cv2.imshow('Camera', frame)

    results = model(frame)

    if results[0].boxes.xyxy.numel() != 0:
        pygame.mixer.Sound.play(fire_sound)

    else:
        pygame.mixer.Sound.stop(fire_sound)

    # when press 'q' key exit the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        pygame.mixer.Sound.stop(fire_sound)
        break

cap.release()
cv2.destroyAllWindows()
```
